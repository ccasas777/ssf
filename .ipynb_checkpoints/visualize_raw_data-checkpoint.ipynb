```python
from matplotlib import pyplot as plt
import random
import numpy as np
```

```python
def parse_txt_2ch(text_lines):
    time, ch1 = [], []
    for i, line in enumerate(text_lines):
        if i < 6:
            continue
        line = line.split("; ")
        try:
            time.append(float(line[0]))
            ch1.append(float(line[1]))
        except:
            print("End line at {} ".format(i))
            break
    return np.asarray(time), np.asarray(ch1)

def parse_txt_3ch(text_lines):
    time, ch1,ch2 = [], [],[]
    for i, line in enumerate(text_lines):
        if i < 7:
            continue
        line = line.split("; ")
        try:
            time.append(float(line[0]))
            ch1.append(float(line[1]))
            ch2.append(float(line[2]))
        except:
            print("End line at {} ".format(i))
            break
    return np.asarray(time), np.asarray(ch1),np.asarray(ch2)

def load_text(path):
    with open(path) as f:
        return [l.replace("\n", "") for l in f.readlines()]
```

```python
data_path = "./raw_data/meas_scope_20210623_122807_3.3umQ37a_75mA_24.50C_cavity20Hz_8Vpp_113M_t_5MHzLP_SM.txt"
time, raw_data = parse_txt_2ch(load_text(data_path))
```

```python
x= np.arange(0,len(raw_data),1)
plt.plot(x,raw_data)
```

```python
start = 1675000
stop = 1700000
plt.plot(x[start:stop],raw_data[start:stop])
```
